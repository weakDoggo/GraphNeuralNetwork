```python
inputs = m.create_input(dataset1)

model_1 = m.create_model(['ginConvo', 'ginConvo', 'ginConvo', 'Pooling', 'Dense'], inputs, units=128, activation="relu")
model_2 = m.create_model(['Convolution','Convolution', 'Pooling', 'Dense'], inputs)
model_3 = m.create_model(['ginConvo', 'Pooling'], inputs)
model_4 = m.create_model(['ginConvo', 'ginConvo', 'ginConvo', 'Pooling', 'Dense'], inputs, units=128, activation="relu", output_units = 128)

callback = tf.keras.callbacks.EarlyStopping(monitor='accuracy', patience=50)
```
